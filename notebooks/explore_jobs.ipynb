```python
# ---------------------------------------------------------
# üõ†Ô∏è 1. Import libraries
# ---------------------------------------------------------
import pandas as pd
import json
```

```python
# ---------------------------------------------------------
# üõ†Ô∏è 2. Load Dataset
# ---------------------------------------------------------
with open("../data/jobs_raw.json", "r", encoding="utf-8") as f:
    data = json.load(f)

df = pd.DataFrame(data)
print(f"Shape: {df.shape}")
df.head()
```

```python
# ---------------------------------------------------------
# üîç 3. Inspect Dataset
# ---------------------------------------------------------
```

```python
# View column names
df.columns.tolist()
```

```python
# View data types and nulls
df.info()
```

```python
# Summary statistics
df.describe(include='all')
```

```python
# Check for missing values
df.isnull().sum().sort_values(ascending=False).head(20)
```

```python
# ---------------------------------------------------------
# üßº 4. Clean Column Names and Remove Irrelevant Ones
# ---------------------------------------------------------
df = df.drop(columns=["company_logo", "url", "tags", "logo", "slug"], errors="ignore")
df = df.rename(columns={
    "company": "Company",
    "position": "Job Title",
    "location": "Location",
    "date": "Posted Date",
    "description": "Description",
    "tags": "Tags",
    "job_type": "Job Type"
})
df.head()
```

```python
# ---------------------------------------------------------
# üìÜ 5. Convert Dates to Datetime
# ---------------------------------------------------------
df["Posted Date"] = pd.to_datetime(df["Posted Date"], errors="coerce")
df["Posted Date"].head()
```

```python
# ---------------------------------------------------------
# üßπ 6. Drop Rows with Missing Key Data
# ---------------------------------------------------------
df = df.dropna(subset=["Company", "Job Title", "Location"])
df.shape
```

```python
# ---------------------------------------------------------
# üåç 7. Normalize 'Location' Field
# ---------------------------------------------------------
df["Location"] = df["Location"].str.strip().str.title()
df["Location"] = df["Location"].replace({
    "Worldwide": "Remote",
    "Anywhere": "Remote",
    "Global": "Remote"
})
df["Location"].value_counts().head(10)
```
